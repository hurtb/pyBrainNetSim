### Simple Display of The Internal Neural Network

```python
%matplotlib inline
import pyBrainNetSim.models.network as network
import pyBrainNetSim.generators.random as random
from pyBrainNetSim.drawing.viewers import draw_networkx
```

```python
sm_prop_dist = random.SensorMoverPropertyDistribution()  # default set of distributions of the different variables in the mouse
net = network.NeuralNetData(sm_prop_dist.create_digraph())
ax = draw_networkx(net)
```

- **SIZE of the NEURONS**: proportional to the total strength of the post-synaptic neurons it synapses onto
- **COLOR of the NEURONS**: Green represents excitatory neurons, and Red inhibitory. The shad of which is proportional to the amount of energy in the neuron.
- **SHAPE of the NEURONS**: represents the type of neuron.
  - Square = Sensory
  - Hexagon = Motor
  - Circle = Internal
- **CONNECTIONS between NEURONS**: indicated by the black lines.
